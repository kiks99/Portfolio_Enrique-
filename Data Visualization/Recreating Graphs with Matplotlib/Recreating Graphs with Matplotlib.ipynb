# Recreating Graphs using Matplotlib!

We are going to imagine that we are working as a high school math teacher. We want to display important metrics about our current class and other classes from previous years, in order to understand patterns and trends that may not be visible from just seeing the numbers in a spreadsheet. 

In this project we are going to re-create some graphs in Matplotlib. Let's go! 

```
#Import Libraries
from matplotlib import pyplot as plt
```

# Bar Chart with Error

Our first graph will be a bar chart of past year's averages with their respective errors of calculation 

```
past_years_averages = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
error = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4]

#Creating our figure and it's size 
plt.figure(figsize= (10,8))

#Creating our subplot for changing axis characteristics
ax = plt.subplot()

#Creating our bar chart with the error and a capsize
plt.bar(range(len(years)), past_years_averages, yerr = error, capsize = 5)

#Changing our axis size
plt.axis([-0.5, 6.5, 70, 95])

# Setting our x ticks and labels 
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years)

#Setting our title, x label and y label
plt.title("Final Exam Averages")
plt.xlabel("Year")
plt.ylabel("Test Average")

#Saving our bar graph as a png file
plt.savefig('my_bar_chart.png')

plt.show()
```

## Side by Side Bars

We are going to create a chart that displays average test scores between students who went to two different middle schools before enrolling in our high school

```
#information needed
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
middle_school_a = [80, 85, 84, 83, 86]
middle_school_b = [73, 78, 77, 82, 86]

#function to create the x-values for the side by side bars
def create_x(t, w, n, d):
    return [t*x + w*n for x in range(d)]

# t represents the number of datasets
# w represents the width of the bar
# n represents the dataset place we are gonna use
# d represents the number of bars we are going to plot

school_a_x = create_x(2, 0.8, 1, 5)
school_b_x = create_x(2, 0.8, 2, 5)

# Creating the figure and it's size
plt.figure(figsize = (10,8))

# Creating the subplot for axis modifications
ax = plt.subplot()

# Plotting the bars
plt.bar(school_a_x, middle_school_a)
plt.bar(school_b_x, middle_school_b)

# We are going to create a list comprehension to put the x tick labels in the middle 
# of both side bars
middle_x = [ (a+b) / 2.0 for a,b in zip(school_a_x, school_b_x)]
ax.set_xticks(middle_x)
ax.set_xticklabels(unit_topics)

# Creating the bar labels, title , x & y labels
plt.legend(['Middle School A', 'Middle School B'])
plt.title("Test Averages on Different Units")
plt.xlabel('Unit')
plt.ylabel('Test Average')

# Saving our figure as a png image
plt.savefig('my_side_by_side.png')

plt.show()
```

## Stacked Bars

We are going to create a stacked chart for the different units for a 30 student group. We'll be stacking the grades by using numpy.  

```
import numpy as np

#Loading data
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]

#Range of bars
x = range(5)

#We'll create the bottoms for stacking the bars by using np.add()
c_bottom = np.add(As, Bs)
d_bottom = np.add(c_bottom, Cs)
f_bottom = np.add(d_bottom, Ds)

#Creating figure
plt.figure(figsize = (10,8))

#Creating ax subplot
ax = plt.subplot()

#Creating bars graph
plt.bar(x, As)
plt.bar(x, Bs, bottom = As)
plt.bar(x, Cs, bottom = c_bottom)
plt.bar(x, Ds, bottom = d_bottom)
plt.bar(x, Fs, bottom = f_bottom)

#Setting x ticks & x tick labels
ax.set_xticks(range(len(unit_topics)))
ax.set_xticklabels(unit_topics)

#Settting legends, title and x&y labels
plt.legend(['A', 'B', 'C', 'D'])
plt.title('Grade Distribution')
plt.xlabel('Unit')
plt.ylabel('Number of students')

#Saving graph as a png
plt.savefig('my_stacked_bar.png')

plt.show()
```

## Labeled Pie Chart

This pie chart will display what students think the hardest topic covered thought out or math course is. 

```
#Load data
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
num_hardest_reported = [1, 3, 10, 15, 1]

#Creating figure
plt.figure(figsize = (10,8))

# Plotting the pie chart
plt.pie(num_hardest_reported, labels = unit_topics, autopct = '%d%%')

#Making the axis equal
plt.axis('equal')

#Making title
plt.title('Hardest Topics')

# Saving fig as a png
plt.savefig('my_pie_chart.png')

plt.show()
```

## Line with Shaded Error

This chart displays the relationship betweeen the time students say they studied and the scores they received on their final exams. 

```
#Load data
hours_reported =[3, 2.5, 2.75, 2.5, 2.75, 3.0, 3.5, 3.25, 3.25,  3.5, 3.5, 3.75, 3.75,4, 4.0, 3.75,  4.0, 4.25, 4.25, 4.5, 4.5, 5.0, 5.25, 5, 5.25, 5.5, 5.5, 5.75, 5.25, 4.75]
exam_scores = [52.53, 59.05, 61.15, 61.72, 62.58, 62.98, 64.99, 67.63, 68.52, 70.29, 71.33, 72.15, 72.67, 73.85, 74.44, 75.62, 76.81, 77.82, 78.16, 78.94, 79.08, 80.31, 80.77, 81.37, 85.13, 85.38, 89.34, 90.75, 97.24, 98.31]

#Creating figure
plt.figure(figsize = (10, 8))

#Creating the lower and upper bouunds for the error of 20%
hours_upper_bound = [element + (element * 0.2) for element in hours_reported]
hours_lower_bound = [element - (element * 0.2) for element in hours_reported]

#Shading the error area on the graph
plt.fill_between(exam_scores, hours_lower_bound, hours_upper_bound, alpha = 0.2)

#Plotting the area
plt.plot(exam_scores, hours_reported, linewidth = 2)

#Creating title, x&y labels
plt.title('Time spent studying vs final exam scores')
plt.xlabel('Score')
plt.ylabel('Hours studying (self-reported)')

#Saving figure as a png image
plt.savefig('my_line_graph.png')

plt.show()
```
