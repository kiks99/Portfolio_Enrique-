Mushrooms exist in a variety of different colors, shapes, sizes, textures, etc. In this project, we will analyze an extensive mushroom dataset from UCI using bar charts and acquaint ourselves with the diverse array of mushrooms that exist worldwide.

```
#Import libraries 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```
#Load the data
df = pd.read_csv("mushroom_data.csv")
print(df.head())
```


Take a look at plot.py where weâ€™ve loaded mushroom_data.csv. It contains 23 columns of data describing thousands of mushrooms. If you read through the header of mushroom_data.csv, you will see that every single variable is a categorical variable. Therefore, we can create bar charts for the counts of every variable in the csv file.

We are going to plot graphs for every variable using a for loop, a list with all of the column headers and the countplot() function

```
# list of all column headers
columns = df.columns.tolist()
print(len(df))
```

```
for column in columns:
  plt.figure()  
  sns.countplot( x = df[column], order = df[column].value_counts().index)
  plt.xticks(rotation = 30, fontsize = 10)
  plt.xlabel(column, fontsize=12)
  plt.title(column + " Value Counts")
  plt.show()
  plt.clf() 
```
