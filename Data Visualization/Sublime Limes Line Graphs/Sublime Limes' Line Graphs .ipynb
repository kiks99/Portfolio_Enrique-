# Sublime Limes' Line Graphs
In this project, we will be acting as a data analyst for an online lime retailer called Sublime Limes. People all over the world can use this product to get the freshest, best-quality limes delivered to their door. One of the product managers at Sublime Limes would like to gain insight into the customers and their sales habits. Using Matplotlib, weâ€™ll create some different line graphs (or maybe, lime graphs) to communicate this information effectively.

```
#Import libraries
from matplotlib import pyplot as plt
```

Let's import the data that we will plot 

```
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
```

```
#Creating a figure with width 12 and height 8
plt.figure(figsize = (12,8))
# We are going to make two charts in one figure, laid out side-by-side. 
ax1 = plt.subplot(1, 2, 1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker = 's')
plt.xlabel('Months')
plt.ylabel('Visits')
plt.title('Visits per Month')
plt.subplots_adjust(bottom = 0.2)
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)
ax2 = plt.subplot(1, 2, 2)
plt.plot(x_values, key_limes_per_month, color = '#808000', marker = 'o', label = 'key limes')
plt.plot(x_values, persian_limes_per_month, color = 'purple', marker = 's', label = 'persina limes')
plt.plot(x_values, blood_limes_per_month, color = 'red', marker = '*', label = 'blood limes')
plt.legend()
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales per Month of Different Limes')
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)
plt.show()
```
